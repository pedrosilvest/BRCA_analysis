#### Plot Coverage by pos negB vs ttdB

```r
#remove other regions where != chr
coverage_neg <- coverage_neg %>% filter(grepl("^chr", chr))
coverage_ttd <- coverage_ttd %>% filter(grepl("^chr", chr))

# Check rows in coverage_neg that are not in coverage_ttd
missing_in_ttd <- coverage_neg %>%
  anti_join(coverage_ttd, by = c("chr", "init_pos", "end_pos"))

# Check rows in coverage_ttd that are not in coverage_neg
missing_in_neg <- coverage_ttd %>%
  anti_join(coverage_neg, by = c("chr", "init_pos", "end_pos"))

# Print results
if (nrow(missing_in_ttd) > 0) {
  cat("Rows in coverage_neg not in coverage_ttd:\n")
  print(missing_in_ttd)
} else {
  cat("All rows in coverage_neg are present in coverage_ttd.\n")
}

if (nrow(missing_in_neg) > 0) {
  cat("Rows in coverage_ttd not in coverage_neg:\n")
  print(missing_in_neg)
} else {
  cat("All rows in coverage_ttd are present in coverage_neg.\n")
}
```

```r
#combine datasets
coverage_combined <- coverage_neg %>%
  inner_join(coverage_ttd, by = c("chr", "init_pos", "end_pos"))
```

```r
print(head(coverage_combined))
```

```r
#prepare data for plotting: Reshape into long format
coverage_long <- coverage_combined %>%
  select(chr, init_pos, end_pos, cov_neg, cov_ttd) %>%
  pivot_longer(
    cols = c("cov_neg", "cov_ttd"),
    names_to = "sample",
    values_to = "coverage"
  )

coverage_long <- coverage_long %>%
  mutate(coverage = coverage + 1) # Add 1 to all coverage values

#facet plot for each chr
# Facet plot with log-transformed coverage
ggplot(coverage_long, aes(x = init_pos, y = coverage, color = sample, group = sample)) +
  geom_line(alpha = 0.8, linewidth = 0.6) +
  facet_wrap(~ chr, scales = "free_y") +
  scale_y_log10() + # Apply log10 scale to the y-axis
  scale_color_manual(values = c("cov_neg" = "blue", "cov_ttd" = "red")) +
  theme_minimal() +
  theme(
    axis.text.x = element_blank(), # Remove x-axis labels
    axis.ticks.x = element_blank(),
    panel.grid.minor = element_blank()
  ) +
  labs(
    x = "Position",
    y = "Log10(Coverage)",
    color = "Sample",
    title = "Log-Transformed Coverage Comparison by Chromosome"
  )

```

```r
coverage_combined <- coverage_combined %>%
  mutate(log2_ratio = log2((cov_ttd + 1) / (cov_neg + 1))) # Add +1 to avoid log(0)

# Plot log2 ratio for chr1
ggplot(coverage_combined %>% filter(chr == "chr1"), aes(x = init_pos, y = log2_ratio)) +
  geom_line(color = "purple", alpha = 0.8, linewidth = 0.6) +
  theme_minimal() +
  labs(
    x = "Position",
    y = "Log2(Tumor/Normal Coverage)",
    title = "Log2 Ratio of Coverage (Tumor vs Normal) for chr1"
  ) +
  geom_hline(yintercept = 0, linetype = "dashed", color = "black")
```

```r
# Scatter plot
ggplot(coverage_combined, aes(x = cov_neg, y = cov_ttd)) +
  geom_point(alpha = 0.5, color = "blue") +
  theme_minimal() +
  labs(
    x = "Normal Coverage",
    y = "Tumor Coverage",
    title = "Coverage Correlation Between Tumor and Normal"
  ) +
  geom_abline(slope = 1, intercept = 0, linetype = "dashed", color = "red") # Line of equality
```

```r

library(zoo)
coverage_combined <- coverage_combined %>%
  filter(chr == "chr1") %>%
  mutate(rolling_log2_ratio = rollmean(log2_ratio, 10, fill = NA))

# Plot rolling averages
ggplot(coverage_combined, aes(x = init_pos, y = rolling_log2_ratio)) +
  geom_line(color = "purple", alpha = 0.8, linewidth = 0.6) +
  theme_minimal() +
  labs(
    x = "Position",
    y = "Rolling Log2(Tumor/Normal Coverage)",
    title = "Smoothed Log2 Ratio of Coverage (Tumor vs Normal) for chr1"
  ) +
  geom_hline(yintercept = 0, linetype = "dashed", color = "black")
```
